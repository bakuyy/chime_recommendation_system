```python
ax = plt.gca()
plt.ylim(0, 150)
plt.xlim(0,1)
sns.countplot(x='instrumentalness', data=songs)
plt.title('Instrumentalness of songs')
plt.show()
```

## Ratings

```python
mean_ratings = songs.groupby('track_name')[['popularity']].mean()
number = 5

lowest_rating = mean_ratings.nsmallest(number, mean_ratings)
lowest_rating #get lowest rated songs

highest_rating = mean_ratings.nlargest(number, mean_ratings)
highest_rating #get highest rated songs
```

## Data Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scalar = MinMaxScaler()

quantitative_df = songs[['duration_ms','danceability','energy','loudness','mode','speechiness','acousticness','instrumentalness','liveness','valence','tempo']]
quantitative_df

fitted_model = scalar.fit(quantitative_df)
fitted_model

scaled_data = fitted_model.transform(quantitative_df)
scaled_data[0:4]
```
